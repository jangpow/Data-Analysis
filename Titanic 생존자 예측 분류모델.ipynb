## 2) 데이터전처리

```python
# train data에 대한 데이터 살펴보기
train_df[train_df['Age'].isnull()] 
```

```python
train_df.isnull().sum()
```

```python
# 'Embarked ' 결측값의 row 살펴보기
#train[train['Embarked'].isnull()]
train_df[train_df['Embarked'].isnull()]
```

```python
# 남성이면서 Pclass가 1인 고객들을 추출하였을 때, 아래와 같은 결과가 나왔으므로 'Embarked'의 결측값은 S로 채워준다.
train_df.loc[(train_df['Sex']=='female')&(train_df['Pclass']==1),'Embarked'].value_counts()
```

```python
train_df.loc[train['Embarked'].isnull(), 'Embarked'] = 'S'
```

```python
# 'Age'에 대한 결측값은 평균값으로 채워준다.
condition=train_df['Age'].mean()
train_df.loc[train['Age'].isnull(), 'Age'] = condition

train_df.info()
```

```python
# 'Cabin'은 머신러닝을 만들 때 크게 필요 없으므로 지워준다. 그 외에 머신러닝에 사용하지 않을 칼럼을 모두 지워준다.

COL_DEL=['PassengerId','Name','Cabin','Ticket']
COL_CAT=['Sex','Embarked']
COL_SD= ['Age','Fare']
COL_NUM=['Pclass','SibSp','Parch']
COL_Y=['Survived']
```

```python
# 범주형 데이터 인코딩 진행 -> 정보의 차이가 유의미하다면 standard, 유의미하지 않다면 one-hot-encoding
train_df_OHE= pd.get_dummies(train_df[COL_CAT])
train_df_OHE
```

```python
# 연속형 데이터 최대최소정규화진행
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
train_df_scaled = scaler.fit_transform(train_df[COL_SD])
train_df_scaled= pd.DataFrame(train_df_scaled , columns=COL_SD)
train_df_scaled
```

```python
# 정규화 된 데이터를 하나의 데이터프레임으로 합치기
#train_df_result=train_df_scaled+train_df_OHE+train[COL_NUM]
train_df_result=pd.concat([train_df_OHE,train_df_scaled,train_df[COL_NUM]],axis=1)
train_df_result
```

## 3) 데이터 모델링

```python
# 훈련데이터와 검증데이터로 데이터 분할
from sklearn.model_selection import train_test_split

X_tr,X_val,y_tr,y_val=train_test_split(train_df_result, train[COL_Y], test_size=0.2, random_state=11 )
```

```python
# 데이터 모델링 
from sklearn.linear_model import SGDClassifier , LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

model1=SGDClassifier()
model2=LogisticRegression()
model3=DecisionTreeClassifier()
model4=RandomForestClassifier()
```

```python
model1.fit(X_tr,y_tr)
model2.fit(X_tr,y_tr)
model3.fit(X_tr,y_tr)
model4.fit(X_tr,y_tr)
```

```python
y_val_pred1 =model1.predict(X_val)
y_val_pred2 =model2.predict(X_val)
y_val_pred3 =model3.predict(X_val)
y_val_pred4 =model4.predict(X_val)
```

```python
# 모델 평가
from sklearn.metrics import accuracy_score
print('SGD 값:{}'.format(accuracy_score(y_val, y_val_pred1 )))
print('Logistic 값:{}'.format(accuracy_score(y_val, y_val_pred2 )))
print('DT 값:{}'.format(accuracy_score(y_val, y_val_pred3 )))
print('RF 값:{}'.format(accuracy_score(y_val, y_val_pred4 )))
```
